```python

# -------------------------------------------------------------
# Topic: Pandas Data Manipulation (Day 1 to Day 3 Combo)
# Author: [তোমার নাম এখানে দিতে পারো]
# Goal: Learning DataFrame Creation, Filtering, and Handling Missing Data
# -------------------------------------------------------------

import pandas as pd
import numpy as np

print("--- Step 1: Creating a DataFrame with Missing Values ---")
# ১. আমরা একটি রিয়েল-ওয়ার্ল্ড ডেমো ডাটা সেট তৈরি করছি যেখানে কিছু ঘর ফাঁকা (None) আছে
data = {
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Headphone"],
    "Price": [45000, 1500, None, 12000, None],  # মাউস ও হেডফোনের দাম missing
    "Quantity": [5, 25, 15, 8, 30]
}

df = pd.DataFrame(data)
print("আসল ডাটা টেবিল (ফাঁকা ঘরসহ):")
print(df)
print("\n" + "="*50 + "\n")


print("--- Step 2: Handling Missing Data (fillna) ---")
# ২. সমাধান: আমরা ধরে নিচ্ছি missing দামগুলোর গড় বা একটি আনুমানিক মান ১৫০০ টাকা দিয়ে পূরণ করব
df["Price"] = df["Price"].fillna(1500)
print("ফাঁকা ঘর ভরাট করার পর টেবিল:")
print(df)
print("\n" + "="*50 + "\n")


print("--- Step 3: Filtering Data based on Condition ---")
# ৩. কন্ডিশন: আমরা শুধু সেই পণ্যগুলো দেখব যেগুলোর Quantity ১৫ বা তার বেশি
high_stock_products = df[df["Quantity"] >= 15]
print("যেসব পণ্যের স্টক (Quantity) ১৫ বা তার বেশি:")
print(high_stock_products)
```
