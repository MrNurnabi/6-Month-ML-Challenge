```python

# -------------------------------------------------------------------------
# Topic: Feature Selection and Train-Test Splitting on Real Dataset
# Author: Nurnabi
# Goal: Selecting specific features and splitting the dataset into Train/Test sets
# -------------------------------------------------------------------------

import pandas as pd
from sklearn.model_selection import train_test_split

print("--- Step 1: Loading Dataset and Selecting Features ---")
# 1. Loading the 1275 laptop dataset
df = pd.read_csv('Laptop_Price_Dataset.csv')

# 2. Selecting specific columns for our Machine Learning model
selected_columns = ["Company", "RAM (GB)", "Price (Euro)", "TypeName"]
selected_df = df[selected_columns]
print("Selected Dataset Head:")
print(selected_df.head())
print("\n" + "="*60 + "\n")


print("--- Step 2: Splitting Dataset into Train and Test Sets ---")
# 3. Splitting data: 80% for training and 20% for testing the model
# test_size=0.2 means 20% data is reserved for evaluation
train_data, test_data = train_test_split(selected_df, test_size=0.2, random_state=42)

# 4. Printing the shapes of the split data
print("Training Data Shape (80%):", train_data.shape)
print("Testing Data Shape (20%):", test_data.shape)
```
