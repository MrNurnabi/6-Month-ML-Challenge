```python

# -------------------------------------------------------------------------
# Topic: Handling Categorical Data using One-Hot Encoding
# Author: Nurnabi
# Goal: Converting text/categorical features into numeric values using pandas
# -------------------------------------------------------------------------

import pandas as pd

print("--- Step 1: Loading Dataset and Selecting Features ---")
# 1. Loading the real-world laptop dataset
df = pd.read_csv('Laptop_Price_Dataset.csv')

# 2. Filtering down to our specific columns
selected_columns = ["Company", "RAM (GB)", "Price (Euro)", "TypeName"]
selected_df = df[selected_columns]
print("Original Dataset with Text Columns:")
print(selected_df.head(3))
print("\n" + "="*60 + "\n")


print("--- Step 2: Applying One-Hot Encoding (pd.get_dummies) ---")
# 3. Converting 'Company' and 'TypeName' text columns into numeric True/False (1/0)
# This allows the Machine Learning model to understand text data smoothly
encoded_df = pd.get_dummies(selected_df, columns=["Company", "TypeName"])

# 4. Displaying the first 3 rows of the newly transformed 27-column dataset
print("Transformed Dataset Ready for Machine Learning:")
print(encoded_df.head(3))
```
